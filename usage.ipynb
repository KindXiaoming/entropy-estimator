```python
from find_projection import *

set_seed()
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'device={device}')

x, labels = get_data()
x, labels = x.to(device), labels.to(device)

# input x: (batch, feature_dim)
# target_dim: project feature_dim to low dimension.
# labels: if labels = None (default), entropy is computed treating all data as one class, 
# otherwise entropy is computed by summing up entropy of each class
# plot_labels: this is only for plotting

x_proj, proj, entropy = find_projection(x, target_dim=2, labels=labels, plot_labels=labels)
```
